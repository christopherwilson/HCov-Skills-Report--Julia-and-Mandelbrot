```python
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
def gen_mandelbrot(tr : ComplexNumber, bl: ComplexNumber, res: tuple[int]) -> tuple[list]:
    """Plotting the Mandelbrot set
    
    tr = the point in the top-left of the figure
    bl = the point in the bottom-right of the figure
    res = the "resolution" of the plot (how many points in each row/column of the plot)
    """
    N = 90 #number of iterations being checked

    #generating points
    start = tr #start point
    diff_bounds = tr-bl
    step_re = (abs(diff_bounds.real()))/res[0]
    step_im = (abs(diff_bounds.imag()))/res[1]
    in_mandelbrot = [
        [], #stores real value of point
        []  #stores imaginary value of point
        ]
    out_mandelbrot = [
        [], # stores real value of point
        [], # stores imaginary value of point
        [] # number of iterations taken to reach 2
        ]
    for i in range(res[0]):
        for j in range(res[0]):
            c = start + CC(i*step_re, -j*step_im)
            bounded = True
            zn = 0
            for itrs in range(1, N+1):
                zn = zn^2 + c #iterating
                if zn.abs() > 2:
                    out_mandelbrot[0].append(c.real())
                    out_mandelbrot[1].append(c.imag())
                    out_mandelbrot[2].append((itrs))
                    bounded = False
                    break 
            
            if bounded:
                in_mandelbrot[0].append(c.real())
                in_mandelbrot[1].append(c.imag())
    return in_mandelbrot, out_mandelbrot

```

```python
_, out_mandelbrot = gen_mandelbrot(CC(-2,1.25), CC(0.5,-1.25), (200, 200))  
plt.scatter(
    out_mandelbrot[0], 
    out_mandelbrot[1], 
    c=out_mandelbrot[2], 
    s = 1,
    cmap= "plasma"
    )
#plt.tight_layout()
ax = plt.gca()
ax.set_aspect("equal")
ax.set_facecolor("black")
ax.margins(x=0, y=0)
plt.xlabel("$\mathrm{Re}(c)$", usetex=True)
plt.ylabel("$\mathrm{Im}(c)$", usetex=True)
```

```python
def gen_julia(tr : ComplexNumber, bl: ComplexNumber, res: tuple[int], c: ComplexNumber, N: int = 20) -> tuple[list]:
    """Plotting the Julia set generated by f(z) = z^2 + c
    
    tr = the point in the top-left of the figure
    bl = the point in the bottom-right of the figure
    res = the "resolution" of the plot (how many points in each row/column of the plot)
    c = start number
    N = max number of iterations
    """

    #generating points
    start = tr #start point
    diff_bounds = tr-bl
    step_re = (abs(diff_bounds.real()))/res[0]
    step_im = (abs(diff_bounds.imag()))/res[1]
    in_julia = [
        [], #stores real value of point
        []  #stores imaginary value of point
        ]
    out_julia = [
        [], # stores real value of point
        [], # stores imaginary value of point
        [] # number of iterations taken to reach 2
        ]
    for i in range(res[0]):
        for j in range(res[0]):
            z0 = start + CC(i*step_re, -j*step_im)
            zn = z0
            bounded = True
            for itrs in range(1, N+1):
                zn = zn^2 + c #iterating
                if zn.abs() > 2:
                    out_julia[0].append(z0.real())
                    out_julia[1].append(z0.imag())
                    out_julia[2].append((itrs))
                    bounded = False
                    break 
            
            if bounded:
                in_julia[0].append(z0.real())
                in_julia[1].append(z0.imag())
    return in_julia, out_julia
```

```python
_, out_julia = gen_julia(CC(-1.5,1.25), CC(1.5,-1.25), (300, 300), CC(-0.12,0.7))
print("plotting...")
fig = plt.scatter(
    out_julia[0], 
    out_julia[1], 
    c=out_julia[2], 
    s = 1,
    cmap= "plasma"
    )
plt.tight_layout()
ax = plt.gca()
ax.set_aspect("equal")
ax.set_facecolor("black")
ax.margins(x=0, y=0)
plt.xlabel("$\mathrm{Re}(z_0)$", usetex=True)
plt.ylabel("$\mathrm{Im}(z_0)$", usetex=True)
plt.title("$c_1 = -0.12+0.7i$")
```
